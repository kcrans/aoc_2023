```python
import math
```

```python
with open('../day8/day8.txt', 'r') as file:
    text = file.read()
```

```python
split_text = text.strip('\n').split('\n')
```

```python
instructs = split_text[0]
instructs
```

```python
network = {}
for line in split_text[2:]:
    source_str, dest_str = line.split(' = ')
    left, right = dest_str.strip('()').split(', ')
    network[source_str] = (left, right)
network    
```

```python
current_node = 'AAA'
steps = 0
while current_node != 'ZZZ':
    for move in instructs:
        steps += 1
        next_ops = network[current_node]
        match move:
            case 'L': 
                current_node = next_ops[0]
            case 'R':
                current_node = next_ops[1]
steps
```

```python
starts = list(filter(lambda x : x[-1] == 'A', network.keys()))
cycles = []
for start in starts:
    steps = 0
    current_node = start
    while True:
        if current_node[-1] == 'Z':
            print(steps, current_node)
        for move in instructs:
            steps += 1
            next_ops = network[current_node]
            match move:
                case 'L': 
                    current_node = next_ops[0]
                case 'R':
                    current_node = next_ops[1]
    print(steps, current_node)
    cycles.append(steps)
```
